# **Global AI Hub Data Analysis Project**

In this file, I will showcase the data visualization project I prepared for the Global AI Hub Data Analysis Bootcamp using the New York City Airbnb Open Data dataset.

```python
# Import necessary libraries

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# Load the dataset and list the first 20 row

df_airbnb = pd.read_csv("AB_NYC_2019.csv")
df_airbnb.head(20)
```

```python
# List the info about dataset file

df_airbnb.info()
```

```python
# Print the column headers of the dataset file

df_airbnb.columns
```

```python
# Example: Histogram of price distribution

plt.figure(figsize=(8, 6))
sns.histplot(df_airbnb['price'], kde=True, bins=100)
plt.title('Price Distribution of NYC Airbnb Listings')
plt.xlabel('Price')
plt.ylabel('Count')
```

```python
# Example: Scatter plot of the relationship between minimum nights and price

plt.figure(figsize=(8,6))
sns.scatterplot(x="minimum_nights", y='price', data=df_airbnb)
```

```python
# Example: KDE plot of density of price

sns.kdeplot(df_airbnb["price"])
```

```python
# Example: Bar plot of average price by neighbourhood group

plt.figure(figsize=(8, 6))
sns.barplot(x='neighbourhood_group', y='price', data=df_airbnb)
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
```

```python
# Example: Box plot of price distribution by neighbourhood group

plt.figure(figsize=(8, 6))
sns.boxplot(x='neighbourhood_group', y='price', data=df_airbnb)
plt.title('Price Distribution by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
```

```python
#Example: Violin Plot of price distribution by room type
plt.figure(figsize=(8, 6))
sns.violinplot(x='room_type', y='price', data=df_airbnb)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
```

```python
# Example: Pie chart of neighbourhood group and price
df_airbnb.groupby("neighbourhood_group").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(8,8))
```

```python
# Example: Bar Plot of neighbourhood group-price-room type

sns.barplot(x="neighbourhood_group",
            y="price",
            hue="room_type",
            data=df_airbnb)
plt.title("Neighbourhood Group - Price - Room Type")
```

```python
# Example: Scatter plot of lattitude by longtitude

plt.figure(figsize=(10, 8))
plt.scatter(df_airbnb['longitude'], df_airbnb['latitude'], alpha=0.5, s=5)  # Adjust 'alpha' and 's' for better visibility
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('NYC Airbnb Listings - Latitude vs. Longitude')
plt.grid(True)
```

```python
# We can even create an interactive scatter plot using Plotly, to have a real map of the dataset
fig = px.scatter_mapbox(df_airbnb, lat='latitude', lon='longitude', color='room_type', hover_name='id', hover_data=['price', 'neighbourhood'],
                        zoom=10)
fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(margin={'r': 0, 't': 0, 'l': 0, 'b': 0})
fig.show()
```
